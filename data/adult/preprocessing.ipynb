```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

adult=pd.read_csv('./adult-census.csv')
adult.head()
```

```python
adult.describe()
```

```python
adult.isna().sum()
```

```python
numerical_columns = [
    "age",
    "education-num",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
]
categorical_columns = [
    "workclass",
    "education",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "native-country",
]
adult['target']=adult['class']
adult.drop(['class'],axis=1,inplace=True)
adult.head()
```

```python
le=LabelEncoder()
for cate in categorical_columns+['target']:
    adult[cate]=le.fit_transform(adult[cate])
adult.head()
```

```python
adult.to_csv('adult_processed.csv',index=False)
```
